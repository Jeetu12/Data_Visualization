### **2. How do you create a basic line plot using Matplotlib?**

**Theory:**
A line plot is useful for visualizing data trends over intervals. You use `plt.plot()` to create one.

**Practical (Python Code):**

```python

plt.ylabel("Y-axis")

plt.show()
```

```python
fig.update_layout(
    title='Custom Line Plot',
    xaxis_title='X Axis',
    yaxis_title='Y Axis',
    template='plotly_dark'
)

```

### **16. Explain when to use a pie chart versus a bar chart.**

**Theory:**

* Use a **pie chart** when you want to show **proportions or percentages** of a whole for a **limited number of categories** (usually less than 6).
* Use a **bar chart** when comparing **absolute values** across **multiple categories**.

**Why prefer bar over pie often?**
Bar charts are easier to read accurately, especially when categories are numerous or values are similar.

**Practical (Pie Chart Example):**

```python
import matplotlib.pyplot as plt

labels = ['A', 'B', 'C']
sizes = [30, 50, 20]

plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title("Pie Chart Example")
plt.show()
```

### **32. What is the purpose of Seaborn's `sns.set_style()` method, and how do you apply different plot styles?**

**Theory:**

* `sns.set_style()` is used to control the **overall aesthetic theme** of Seaborn plots.
* It helps make plots more readable and professional.
* Available styles:

  * `"darkgrid"` (default, great for regression)
  * `"whitegrid"` (good for plots with categorical data)
  * `"dark"`
  * `"white"`
  * `"ticks"`

**Practical:**

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")
sns.boxplot(x="day", y="total_bill", data=tips)
plt.title("Boxplot with Seaborn Style")
plt.show()
```
